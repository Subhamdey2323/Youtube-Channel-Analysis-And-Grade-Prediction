## 9. Data Cleaning [ Video Uploads & Subscribers ]

```python
# Replace '--' with NaN
data['Video Uploads'] = data['Video Uploads'].replace('--', np.nan)

# Convert the column to integers
data['Video Uploads'] = data['Video Uploads'].astype(float).astype('Int32')
```

```python
# Replace NaN with 0
data['Video Uploads'] = data['Video Uploads'].fillna(0)

# Convert the column to integers
data['Video Uploads'] = data['Video Uploads'].astype(int)
```

```python
data.dtypes
```

```python
# Replace '--' with NaN
data['Subscribers'] = data['Subscribers'].replace('--', np.nan)
```

```python
data['Subscribers'] = data['Subscribers'].fillna(0)
```

```python
# Replace non-numeric values with NaN
data['Subscribers'] = pd.to_numeric(data['Subscribers'], errors='coerce')

# Convert the column to integers
data['Subscribers'] = data['Subscribers'].astype('Int32')
```

## 19.Feature Engineering

```python
# Encode categorical variables
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Grade_encoded'] = le.fit_transform(data['Grade'])
```

```python
data
```

## 20. Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split

# Feature & target selection
features = ['Video Uploads', 'Subscribers', 'Video views']
X = data[features]
y = data['Grade_encoded']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=500)
log_reg.fit(X_train, y_train)
y_pred_log = log_reg.predict(X_test)
```

### Decision-Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train, y_train)
y_pred_dt = dt_clf.predict(X_test)
```

### Random-Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X_train, y_train)
y_pred_rf = rf_clf.predict(X_test)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
```

### Support Vector Classifier

```python
from sklearn.svm import SVC

svm_clf = SVC()
svm_clf.fit(X_train, y_train)
y_pred_svm = svm_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

models = {
    "Logistic Regression": log_reg,
    "Decision Tree": dt_clf,
    "Random Forest": rf_clf,
    "KNN": knn,
    "SVM": svm_clf
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"📌 {name}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred, target_names=le.classes_))
    print("-"*50)
```

```python
from sklearn.model_selection import cross_val_score

for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5)
    print(f"{name} - Average CV Accuracy: {scores.mean():.4f}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = GradientBoostingClassifier()
gb_clf.fit(X_train, y_train)
y_pred_gb = gb_clf.predict(X_test)

print("Gradient Boosting Classifier")
print("Accuracy:", accuracy_score(y_test, y_pred_gb))
print(classification_report(y_test, y_pred_gb, target_names=le.classes_))
```

```python
accuracies = []

for name, model in models.items():
    acc = accuracy_score(y_test, model.predict(X_test))
    accuracies.append((name, acc))

accuracies.append(("Gradient Boosting", accuracy_score(y_test, y_pred_gb)))

# Convert to DataFrame
acc_df = pd.DataFrame(accuracies, columns=['Model', 'Accuracy'])

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x='Model', y='Accuracy', data=acc_df, palette='pastel')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.title("Model Accuracy Comparison")
plt.tight_layout()
plt.show()
```
